### Life expectancy and Healthy life expectancy
lifeExpectancyAtBirth.csv -> Life expectancy at birth, country wise mentioned in age (years).

HALElifeExpectancyAtBirth.csv -> Healthy life expectancy (HALE) at birth, country wise mentioned in age(years).

WHOregionLifeExpectancAtBirth.csv -> Life expectancy at birth, Region wise mentioned in age (years).

HAleWHOregionLifeExpectancy.csv -> Healthy life expectancy at birth, region wise mentioned in age(years).

%HaleInLifeExpectancy.csv -> Healthy life and life expectancy at birth with the % of HALE in life expectancy.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import ticker

# load dataset and initialize
lifeExpectancyAtBirth = pd.read_csv("lifeExpectancyAtBirth.csv")
HALElifeExpectancyAtBirth = pd.read_csv("HALElifeExpectancyAtBirth.csv")
WHOregionLifeExpectancyAtBirth = pd.read_csv("WHOregionLifeExpectancyAtBirth.csv")
HALeWHOregionLifeExpectancyAtBirth = pd.read_csv("HALeWHOregionLifeExpectancyAtBirth.csv")

# rename column
HALeWHOregionLifeExpectancyAtBirth.rename(columns={"Hale Expectency": "First Tooltip"}, inplace=True)

# create name attribute
lifeExpectancyAtBirth.name = ("National Life Expectancy at Birth")
HALElifeExpectancyAtBirth.name = ("National Healthy Life Expectancy at Birth")
WHOregionLifeExpectancyAtBirth.name = ("Regional Life Expectancy at Birth")
HALeWHOregionLifeExpectancyAtBirth.name = ("Regional Healthy Life Expectancy at Birth")
```

```python
# lineplot
def lineplot(df, location):
    mask = df["Location"] == location
    df_location = df[mask]
    sns.lineplot(data=df_location, x='Period', y='First Tooltip', hue = "Dim1")
    plt.gca().xaxis.set_major_formatter(ticker.StrMethodFormatter("{x:.0f}"))
    plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:.2f}"))
    plt.xlabel("Time")
    plt.ylabel("Life Expectancy")
    plt.legend(loc = "lower right")
    plt.title(f"{df.name} in {location} over Time")
    plt.style.use('fivethirtyeight')
    plt.show()
    
lineplot(HALeWHOregionLifeExpectancyAtBirth, "Africa")
```

```python
# scatterplot
def scatterplot(df, location):
    mask = df["Location"] == location
    df_location = df[mask]
    sns.scatterplot(data=df_location, x='Period', y='First Tooltip', hue = "Dim1")
    plt.gca().xaxis.set_major_formatter(ticker.StrMethodFormatter("{x:.0f}"))
    plt.gca().yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:.2f}"))
    plt.xlabel("Time")
    plt.ylabel("Life Expectancy")
    plt.legend(loc = "lower right")
    plt.title(f"{df.name} in {location} over Time")
    plt.style.use('fivethirtyeight')
    plt.show()
    
scatterplot(HALeWHOregionLifeExpectancyAtBirth, "Africa")
```

```python
# barplot
def barplot(df, location):
    mask = df["Location"] == location
    df_location = df[mask]
    sns.barplot(data=df_location, x='Dim1', y = "First Tooltip", hue = "Period")
    plt.xlabel("Time")
    plt.ylabel("Life Expectancy")
    plt.legend(loc = "lower right")
    plt.title(f"{df.name} in {location} over Time")
    plt.style.use('fivethirtyeight')
    plt.show()
    
barplot(HALeWHOregionLifeExpectancyAtBirth, "Africa")
```
